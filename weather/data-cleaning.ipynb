```python
import pandas as pd
```

```python
df = pd.read_csv('sf-weather.csv')
display(df)
display(df.info())
```

```python
def extract_int(string):
    return int(string.split(' ')[0])

def extract_float(string):
    return float(string.split(' ')[0])

df['Temperature (F)'] = df['Temperature (F)'].apply(extract_int)
df['Humdity'] = df['Humdity'].apply(extract_int)
df['Wind Speed (mph)'] = df['Wind Speed (mph)'].apply(extract_int)
df['Rainfall (in)'] = df['Rainfall (in)'].apply(extract_float)
```

```python
df.info()
```

```python


df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df.info()
```

```python
import datetime

def to_correct_date(date, time):
    return datetime.datetime.combine(date.date(), time.time())

df.insert(0, 'Timestamp', df['Date'].combine(df['Time'], to_correct_date))
df = df.drop(['Date', 'Time'], axis = 1)
```

```python
print(f'Targeted size: {24 * 61}')
df = df[df['Timestamp'].dt.minute == 56]
df
```

```python
def convert_to_multiple(timestamp):
    base_time = timestamp - datetime.timedelta(minutes = 1)
    return [base_time - datetime.timedelta(minutes = 25), 
            base_time - datetime.timedelta(minutes = 20), 
            base_time - datetime.timedelta(minutes = 15),
            base_time - datetime.timedelta(minutes = 10), 
            base_time - datetime.timedelta(minutes = 5), 
            base_time, 
            base_time + datetime.timedelta(minutes = 5),
            base_time + datetime.timedelta(minutes = 10),
            base_time + datetime.timedelta(minutes = 15),
            base_time + datetime.timedelta(minutes = 20),
            base_time + datetime.timedelta(minutes = 25),
            base_time + datetime.timedelta(minutes = 30)]

df['Timestamp'] = df['Timestamp'].apply(convert_to_multiple)
df = df.explode('Timestamp')
df
```

```python
start_date = datetime.datetime(2018, 1, 1)
end_date = datetime.datetime(2018, 3, 1)
print(start_date)
print(end_date)
df = df[(df['Timestamp'].dt.date >= start_date.date()) & (df['Timestamp'].dt.date < end_date.date())]
df.reset_index(drop = True, inplace = True)
df
```

```python
def split_into(string):
    return string.split(' / ')
df['Condition'].apply(split_into)
df = df.join(pd.get_dummies(df['Condition'].apply(split_into).explode()).groupby(level=0).sum())
df = df.drop('Condition', axis = 1)
df
```

```python
df = df.rename(columns = {'Humdity' : 'Humidity', 'Temperature (F)' : 'Temp (F)'})
df.to_csv('sf-weather-cleaned.csv')
```
